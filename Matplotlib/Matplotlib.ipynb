```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

# Line plot

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
##plt plot 
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
plt.plot(x,y,'r',linestyle='dashed',marker='*',markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(y,"g",marker='+',markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph Diagram')
```

# Bar plot

```python
x=[2,8,10]
y=[11,6,9]



x2=[3,9,11]
y2=[5,16,7]

plt.bar(x,y)
plt.bar(x2,y2)
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
x=[2,8,10]
y=[11,6,9]



x2=[3,9,11]
y2=[5,16,7]

plt.barh(x,y)
plt.bar(x2,y2)
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# Hist Plot

```python
a=np.array([22,67,35,45,56,12,67,89,90])
plt.hist(a)
plt.title('Hist plot')
plt.show()
```

# Box plot

```python
data=[np.random.normal(0,std,100)for std in range(1,4)]
data
```

```python
plt.boxplot(data)
plt.show()
```

# Pie Plot

```python
size=[215,246,130,234]
labels=['python','C','Java','Ruby']
explode=[0.04,0.04,0.2,0.04]
colors=['cyan','orange','pink','yellow']
plt.pie(size,labels=labels,explode=explode,autopct='%1.2f%%',colors=colors,shadow =6,counterclock=False,rotatelabels=True)
plt.show()
```

# Functional Method

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
```

```python
plt.subplot(1,2,1) # rowsColumsPosition
plt.plot(x,y,'r')

plt.subplot(1,2,2) # rowsColumsPosition
plt.plot(x,y,'g')
```

```python
plt.subplot(2,2,1) # rowsColumsPosition
plt.plot(x,y,'r',marker='*',markersize=5)

plt.subplot(2,2,2) # rowsColumsPosition
plt.plot(x,y,'y',marker='*',markersize=5)

plt.subplot(2,2,3) # rowsColumsPosition
plt.plot(x,y,'g',marker='*',markersize=5)

plt.subplot(2,2,4) # rowsColumsPosition
plt.plot(x,y,'b',marker='*',markersize=5)
```

```python
plt.subplot(3,2,1) # rowsColumsPosition
plt.plot(x,y,'r',marker='*',markersize=5)

plt.subplot(3,2,2) # rowsColumsPosition
plt.plot(x,y,'y',marker='*',markersize=5)

plt.subplot(3,2,3) # rowsColumsPosition
plt.plot(x,y,'g',marker='*',markersize=5)

plt.subplot(3,2,4) # rowsColumsPosition
plt.plot(x,y,'b',marker='*',markersize=5)

plt.subplot(3,2,5) # rowsColumsPosition
plt.plot(x,y,'cyan',marker='*',markersize=5)

plt.subplot(3,2,6) # rowsColumsPosition
plt.plot(x,y,'pink',marker='*',markersize=5)
```

# Object oriented method

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.2,0.5,0.5]) ##add_axes(left,bottom,width,height)

axes.plot(x,y)
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.2,0.7,0.8]) ##add_axes(left,bottom,width,height)
axes1.plot(x,y,'r',label='red')

axes2=fig.add_axes([0.2,0.6,0.2,0.3]) ##add_axes(left,bottom,width,height)
axes2.plot(x,y,'b',label='blue')

axes1.legend(loc=4)
axes2.legend()
```
